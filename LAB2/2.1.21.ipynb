```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def f(x):
    return np.power(np.cos(x),2) - np.cos(x)/2 + 1/18

x = np.arange(1,1.75,0.1)
y = f(x)
plt.plot(x,y)
plt.plot(x,x*0,'k')
plt.grid(True)
plt.show()
```

```python
def bisec(f, a, b, eps):
    an = a
    bn = b
    k = 0
    while(bn - an)/2 > eps:
        xn = (an + bn)/2
        fa = f(an)
        fb = f(bn)
        fxn = f(xn)
        if fa*fxn <= 0:
            bn = xn
        else:
            an = xn
        k = k + 1
    xn = (an + bn)/2
    return xn, k
print('x1 = %.10f, iterations = %d' %bisec(f, 0, 1.3, 10**(-10)))
print('x1 = %.10f, iterations = %d' %bisec(f, 1.3, 2, 10**(-10)))
```

```python
root = optimize.root(f, [0.2, 1.9]).x.tolist()
print('x1 = %.10f' %(float(root[0])))
print('x2 = %.10f' %(float(root[1])))
```

```python
def g(x):
    return np.cos(x)**2 - np.cos(x)*2/3 + 1/9
x = np.arange(1,1.75,0.1)
y = g(x)
plt.plot(x,y)
plt.plot(x,x*0,'k')
plt.grid(True)
plt.show()
```

```python
print('x1 = %.10f, iterations = %d' %bisec(g, 1, 1.5, 10**(-10)))
```

```python
root = optimize.root(g, [0.1, 2.1]).x.tolist()
print('x = %.10f' %(float(root[0])))
```
